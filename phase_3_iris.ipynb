```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import load_iris # contains sample datasets like iris, digits, breast_cancer, etc.
from sklearn.model_selection import train_test_split # splits data arrays into random training and testing subsets.
```

```python
data = load_iris() # loading data from sklearn datasets returns a Bunch object (a dictionary-like object provided by scikit-learn)

'''
It contains:

data.data → feature matrix (shape (150, 4))

data.target → class labels (shape (150,))

data.feature_names → names of the 4 features

data.target_names → names of the 3 classes ('setosa', 'versicolor', 'virginica')

This dataset has 150 samples, each describing a flower using 4 numeric measurements:

sepal length (cm)

sepal width (cm)

petal length (cm)

petal width (cm)
'''

X = data.data # nd-array containg data
y = data.target #nd-array containg target
fn = data.feature_names #nd-array containing feature-names
tn = data.target_names #nd-array containing target-names
```

```python
print(f"Shape of X is {X.shape}")
print(f"Shape of y is {y.shape}")
print(fn)
print(tn)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2)
```

### Visualizing the split using numpy and matplotlib

```python
counts = np.bincount(y_train)
position = np.arange(3)
plt.bar(position, counts)
plt.xticks(position, data.target_names)
plt.title('Distribution of Training data')
```
