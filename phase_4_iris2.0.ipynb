```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.model_selection import StratifiedShuffleSplit
```

```python
# loading iris dataset
data = load_iris()
```

```python
X = data.data
y = data.target
```

```python
train_X, test_X, train_y, test_y = train_test_split(X, y, test_size = 0.2)
```

```python
c1 = np.bincount(train_y) # number of unique target classes
p1 = np.arange(len(c1))
```

```python
plt.bar(p1, c1)
plt.xticks(p1, data.target_names)
```

```python
c2 = np.bincount(test_y) # number of unique target classes
p2 = np.arange(len(c2))
plt.bar(p2, c2)
plt.xticks(p2, data.target_names)
```

## We can see that the above two distributions are not identical; thus we can't use train_test_split() method od sklearn.model_selection package

```python
split = StratifiedShuffleSplit(n_splits = 1, test_size = 0.2)

for train_idx, test_idx in split.split(X, y):
    X_train, X_test = X[train_idx], X[test_idx]
    y_train, y_test = y[train_idx], y[test_idx]

c1 = np.bincount(y_train) # number of unique target classes
p1 = np.arange(len(c1))
plt.bar(p1, c1)
plt.xticks(p1, data.target_names)
```

```python
c2 = np.bincount(y_test) # number of unique target classes
p2 = np.arange(len(c2))
plt.bar(p2, c2)
plt.xticks(p2, data.target_names)
```
